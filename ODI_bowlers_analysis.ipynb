###8. Players played for ICC





```python
df[['Player_Name','Country']]=df['Player'].str.split('(', expand=True)
df.head(1)
```

```python
#drop the Player coloumn
df=df.drop('Player',axis=1)
```

```python
#remove ')' from  country

df['Country']=df['Country'].str.replace(")", "")
df.head(1)
```

```python
def icc_check(x):
  if "ICC" in x:
    return "Yes"
  else:
    return "No"

def Asia_check(x):
  if "Asia" in x:
    return "Yes"
  else:
    return "No"

def Afr_check(x):
  if "Afr" in x:
    return "Yes"
  else:
    return "No"
```

```python
df['Played_for_ICC']=df['Country'].apply(icc_check)
df['Played_for_Afr']=df['Country'].apply(Afr_check)
df['Played_for_Asia']=df['Country'].apply(Asia_check)

df.head(1)
```

```python
 #Players Played for  ICC
df.Played_for_ICC.value_counts()
```

Removing "ICC/",'Afr/' from the 'Country'

```python
df['Country']=df['Country'].str.replace('Afr/','')
df['Country']=df['Country'].str.replace('ICC/','')
df['Country']=df['Country'].str.replace('/ICC','')
df['Country']=df['Country'].str.replace('Asia/','')
```

###10.  player(s) had played for the longest period of time


split the span

```python
df[['Start_year','End_year']]=df['Span'].str.split('-',expand=True)
df.head(3)
```

```python
df=df.drop('Span',axis=1)# drop the span column
```

```python
df.info()
```

```python
#convert 'object' data into 'int' data
df[['Start_year','End_year']]=df[['Start_year','End_year']].astype('int64')
```

```python
# find total years played by a player
df['years_played']=df['End_year']-df['Start_year'] +1
```

```python
#drop start_year,end_year
df=df.drop(['Start_year','End_year'],axis=1)
df.head(3)
```

```python
df.columns
```

```python
#Re-arrange  column 
new_col_seqenc=['Player_Name', 'Country','Matches', 'Innings', 'Balls_blowed', 'Runs_conceded', 'Wkts',
       'Average_blow', 'Economy_rate', 'strike_rate', 'Four_wkts', 'Five_wkts',
        'Played_for_ICC', 'Played_for_Afr',
       'Played_for_Asia', 'years_played']
df=df[new_col_seqenc]
```
